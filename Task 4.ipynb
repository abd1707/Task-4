```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('financial_year_survey.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
ratio=['Year']
for values in ratio:
    df[values].plot(kind='hist',bins=10)
    plt.title('{valname} percentage'.format(valname=values))
    plt.xlabel('Given criteria')
    plt.show()
```

```python
value_sort=df.sort_values(by='Value', ascending=False)
plt.figure(figsize=(10,10))
sns.barplot(data=value_sort, x='Variable_name', y='Year')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(data=df, x='Variable_name', y='Year', kind="bar", height=4, aspect=3)
plt.xticks(rotation=90)
plt.show()
```
